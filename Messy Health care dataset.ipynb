```python
import pandas as pd
import numpy as np
```

```python
#Loading the data and checking if it is correctly loaded
df=pd.read_csv(r'C:\Users\HP\OneDrive\Dokumente\Health care messy data.csv')
df.head(5)
```

```python
# checking the data information of the dataset
df.info()
```

# Data cleaning process

```python
# Removing the spaces and unnecssary character from the dataset
for column in df.select_dtypes(include=['object']).columns:
    df[column] = df[column].str.strip()
```

```python
df.head(5)
```

```python
#Replacing forty to 40 in the age columns
df.Age=df.Age.replace("forty",40)
# Handling non numerical value ,changing the datatype of the age to int and also to fill the NAN to be zero(0)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Age'] = df['Age'].replace([np.inf, -np.inf], np.nan)
df['Age'] = df['Age'].fillna(0).astype('int64')
```

```python
df.Age
```

```python
df['Visit Date'] = pd.to_datetime(df['Visit Date'], errors='coerce')
```

```python
#Handling non numerical value ,changing the datatype of the Cholesterol to int 
df["Cholesterol"]=pd.to_numeric(df["Cholesterol"],errors="coerce")
df["Cholesterol"]=df["Cholesterol"].fillna(0).astype("int64")
```

```python
df["Cholesterol"]
```

```python
# Correct the Blood Pressure column 
df['Blood Pressure'] = df['Blood Pressure'].replace('NaN', np.nan)
```

```python
#Normalize text in Medication Name and patient Name
df['Medication'] = df['Medication'].str.title()
df['Patient Name'] = df['Patient Name'].str.title()
df.head(5)
```

```python
#Fill NaN values for numerical columns with the mean
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Cholesterol'].fillna(df['Cholesterol'].mean(), inplace=True)
```

```python
# Replace empty or missing values in columns that contain words or categories  
#(like Gender or Condition) with either a default word (like "Unknown") or the most common value in that column.
df['Gender'].fillna('Unknown', inplace=True)
df['Condition'].fillna(df['Condition'].mode()[0], inplace=True)
df['Medication'].fillna(df['Medication'].mode()[0], inplace=True)
df['Blood Pressure'].fillna('120/80', inplace=True)
df['Email'].fillna('no_email@example.com', inplace=True)
df['Phone Number'].fillna('No_Phone', inplace=True)
```
